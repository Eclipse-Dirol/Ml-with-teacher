```python
df['quatile'] = pd.qcut(x=df['LIMIT_BAL'], q=4, labels=False)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from scipy.stats import uniform, randint, loguniform
```

```python
X = df
y = target
```

```python
df.info()
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
cat_cols = df[["SEX", "EDUCATION", "MARRIAGE"]].columns.tolist()
num_cols = [col for col in X.columns if col not in cat_cols]
```

```python
prep_logreg = ColumnTransformer(transformers=[
    ("cat_cols", OneHotEncoder(handle_unknown='ignore'), cat_cols),
    ("num_cols", StandardScaler(), num_cols)
])
```

```python
pipe_logreg = Pipeline(steps=[
    ('prep', prep_logreg),
    ('logreg', LogisticRegression())
])
```

```python
param_logreg = {
    'logreg__C': loguniform(1e-4, 10),
    'logreg__penalty': ['l1', 'l2'],
    'logreg__solver': ['saga', 'liblinear'],
    'logreg__class_weight': ['balanced', None]
}

rscv_logreg = RandomizedSearchCV(
    estimator=pipe_logreg,
    param_distributions=param_logreg,
    n_iter=40,
    scoring='roc_auc',
    cv=cv,
    n_jobs=-1,
    random_state=42,
    verbose=1
)

rscv_logreg.fit(X,y)

print(rscv_logreg.best_score_)
print(rscv_logreg.best_params_)
```

```python
# 0.7251555325841469
param_for_logreg = {'logreg__C': np.float64(7.39226614051604), 'logreg__class_weight': 'balanced', 'logreg__penalty': 'l2', 'logreg__solver': 'liblinear'}
```

```python
prep_rf = ColumnTransformer(transformers=[
    ('cat_cols', OneHotEncoder(handle_unknown='ignore'), cat_cols),
    ('num_cols', 'passthrough', num_cols)
])
```

```python
pipe_rf = Pipeline(steps=[
    ('prep', prep_rf),
    ('rf', RandomForestClassifier())
])
```

```python
param_rf = {
    'rf__n_estimators': randint(150, 250),
    'rf__max_depth': randint(5,25),
    'rf__min_samples_split': randint(2,30),
    'rf__criterion': ['gini', 'entropy', 'mse', None],
    'rf__max_features': ['sqrt', 0.7, 0.9]
}

rscv_rf = RandomizedSearchCV(
    estimator=pipe_rf,
    param_distributions=param_rf,
    n_iter=20,
    scoring='roc_auc',
    cv=cv,
    n_jobs=-1,
    random_state=42,
    verbose=1
)

rscv_rf.fit(X,y)

print(rscv_rf.best_score_)
print(rscv_rf.best_params_)
```

```python
# 0.7825749795413229
param_for_rf = {'rf__criterion': 'entropy', 'rf__max_depth': 10, 'rf__max_features': 0.7, 'rf__min_samples_split': 11, 'rf__n_estimators': 153}
```

```python
best_param_logreg = {k.replace('logreg__', ''): v for k,v in param_for_logreg.items()}
best_param_rf = {k.replace('rf__', ''): v for k,v in param_for_rf.items()}
```

```python
ensemble = VotingClassifier(estimators=[
    ('logreg', Pipeline(steps=[
        ('prep', prep_logreg),
        ('logreg', LogisticRegression(**best_param_logreg))
    ])),
    ('rf', Pipeline(steps=[
        ('prep', prep_rf),
        ('rf', RandomForestClassifier(**best_param_rf))
    ]))],
    voting='soft',
    n_jobs=-1,
    verbose=3
)

ensemble.fit(X,y)
```

```python
cross_val_score(
    estimator=ensemble,
    X=X,
    y=y,
    scoring='roc_auc',
    cv=cv
)
```

```python
cross_val_score(
    estimator=pipe_rf,
    X=X,
    y=y,
    scoring='roc_auc',
    cv=cv
)
```

```python
pipe_best_rf = Pipeline(steps=[
    ('prep', prep_rf),
    ('rf', RandomForestClassifier(**best_param_rf))
])
```

```python
cross_val_score(
    estimator=pipe_best_rf,
    X=X,
    y=y,
    scoring='roc_auc',
    cv=cv
)
```
